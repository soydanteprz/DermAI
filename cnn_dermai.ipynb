# Classification of dermatoscopic images using Convolutional Neural Networks (CNNs)



Main library imports

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import optimizers, models, layers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.applications import MobileNetV2
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import time

# ========================================
# MAC M3 PRO GPU SETUP
# ========================================

def setup_gpu():
    """Configure TensorFlow for GPU acceleration if available"""

    gpus = tf.config.list_physical_devices('GPU')
    gpu_available = len(gpus) > 0

    if gpu_available:
        try:
            # Enable memory growth
            for gpu in gpus:
                tf.config.experimental.set_memory_growth(gpu, True)

            # Mixed precision for better performance
            policy = tf.keras.mixed_precision.Policy('mixed_float16')
            tf.keras.mixed_precision.set_global_policy(policy)

            print("GPU acceleration enabled")
        except Exception as e:
            print(f"GPU setup failed: {e}")
            gpu_available = False
    else:
        print("ğŸ’» Using CPU (no GPU detected)")

    return gpu_available


print("=" * 70)
print("6-class classification on balanced dataset")
print("Expected: 500 samples per class")
print("ğŸ–¼Image size: 150Ã—150 pixels")
print()

# Set up GPU before importing other modules
gpu_available = setup_gpu()

# Define paths
base_dir = os.path.join(os.getcwd(), 'data')
balanced_train_dir = os.path.join(base_dir, 'balanced_train')
test_dir = os.path.join(base_dir, 'test')
model_save_dir = os.path.join('models')
validation_dir = os.path.join(base_dir, 'validation')

# Create a models directory if it doesn't exist
os.makedirs(model_save_dir, exist_ok=True)

def verify_dataset():
    """Verify the balanced dataset and validation folder exist and check their structure"""

    print("VERIFYING DATASET...")
    print("-" * 40)

    if not os.path.exists(balanced_train_dir):
        raise FileNotFoundError("Balanced training dataset not found. Please run augmentation_script.py first.")

    if not os.path.exists(validation_dir):
        raise FileNotFoundError("Validation dataset not found. Please create a validation set.")

    if not os.path.exists(test_dir):
        raise FileNotFoundError("Test dataset not found. Please create a test set.")

    # Count classes and samples in each directory
    train_classes = [d for d in os.listdir(balanced_train_dir)
                     if os.path.isdir(os.path.join(balanced_train_dir, d))]
    val_classes = [d for d in os.listdir(validation_dir)
                   if os.path.isdir(os.path.join(validation_dir, d))]
    test_classes = [d for d in os.listdir(test_dir)
                    if os.path.isdir(os.path.join(test_dir, d))]

    print("Dataset structure verified:")
    print(f"Training classes: {len(train_classes)} - {sorted(train_classes)}")
    print(f"Validation classes: {len(val_classes)} - {sorted(val_classes)}")
    print(f"Test classes: {len(test_classes)} - {sorted(test_classes)}")

    # Count samples in each directory
    total_train = 0
    total_val = 0
    total_test = 0

    for class_name in sorted(train_classes):
        train_count = len([f for f in os.listdir(os.path.join(balanced_train_dir, class_name))
                           if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
        val_count = len([f for f in os.listdir(os.path.join(validation_dir, class_name))
                         if f.lower().endswith(('.png', '.jpg', '.jpeg'))]) if class_name in val_classes else 0
        test_count = len([f for f in os.listdir(os.path.join(test_dir, class_name))
                          if f.lower().endswith(('.png', '.jpg', '.jpeg'))]) if class_name in test_classes else 0

        total_train += train_count
        total_val += val_count
        total_test += test_count

        print(f"{class_name}: Train={train_count}, Val={val_count}, Test={test_count}")

    print(f"ğŸ“Š TOTAL SAMPLES:")
    print(f"Training: {total_train} samples")
    print(f"Validation: {total_val} samples")
    print(f"Test: {total_test} samples")
    print()
    return True

def create_data_generators():
    """Create optimized data generators using your separate train/validation/test folders"""

    print("CREATING DATA GENERATORS (USING SEPARATE FOLDERS)...")
    print("-" * 40)

    # Optimized batch size for Mac M3 Pro GPU
    batch_size = 32
    target_size = (96, 96)

    # Training data generator WITH augmentation (only for training data)
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        horizontal_flip=True,
        rotation_range=15,
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=0.1,
        zoom_range=0.1,
        fill_mode='nearest'
    )


    eval_datagen = ImageDataGenerator(rescale=1./255)


    train_generator = train_datagen.flow_from_directory(
        balanced_train_dir,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=True,
        seed=42
    )

    validation_generator = eval_datagen.flow_from_directory(
        validation_dir,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=False,
        seed=42
    )

    test_generator = eval_datagen.flow_from_directory(
        test_dir,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=False
    )

    print(f"Training samples: {train_generator.samples} (from balanced_train/)")
    print(f"Validation samples: {validation_generator.samples} (from validation/)")
    print(f"Test samples: {test_generator.samples} (from test/)")
    print(f"Batch size: {batch_size} (optimized for {'GPU' if gpu_available else 'CPU'})")
    print(f"Number of classes: {len(train_generator.class_indices)}")
    print(f"Class indices: {train_generator.class_indices}")
    print()

    # Verify class indices match across all generators
    if (train_generator.class_indices == validation_generator.class_indices == test_generator.class_indices):
        print("Class indices consistent across all datasets")
    else:
        print(f"     Train: {train_generator.class_indices}")
        print(f"     Val: {validation_generator.class_indices}")
        print(f"     Test: {test_generator.class_indices}")

    return train_generator, validation_generator, test_generator

# python
def create_model(model_type, num_classes=6, input_shape=(96, 96, 3)):
    """
    Create CNN model for skin cancer classification based on the paper's architectures.
    """
    print(f"\nCREATING {model_type.upper()} MODEL...")
    print("-" * 40)

    if model_type == 'custom':
        # Custom CNN Architecture
        model = models.Sequential([
            layers.Input(shape=input_shape),
            layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
            layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
            layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
            layers.MaxPooling2D((2, 2)),
            layers.BatchNormalization(),
            layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
            layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
            layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
            layers.MaxPooling2D((2, 2)),
            layers.BatchNormalization(),
            layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
            layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
            layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
            layers.MaxPooling2D((2, 2)),
            layers.BatchNormalization(),
            layers.GlobalAveragePooling2D(),
            layers.Dense(256, activation='relu'),
            layers.Dropout(0.5),
            layers.Dense(128, activation='relu'),
            layers.Dropout(0.5),
            layers.Dense(num_classes, activation='softmax')
        ])
        optimizer = optimizers.SGD(learning_rate=0.0001, momentum=0.9)

    elif model_type == 'mobilenet':
        base_model = MobileNetV2(
            input_shape=input_shape,
            include_top=False,
            weights='imagenet'
        )
        for layer in base_model.layers[:-23]:
            layer.trainable = False
        trainable_count = sum([1 for layer in base_model.layers if layer.trainable])
        non_trainable_count = len(base_model.layers) - trainable_count
        print(f"Base model layers: {len(base_model.layers)}")
        print(f"Trainable layers: {trainable_count}")
        print(f"Frozen layers: {non_trainable_count}")
        inputs = layers.Input(shape=input_shape)
        x = base_model(inputs, training=True)
        x = layers.GlobalAveragePooling2D()(x)
        x = layers.Dense(256, activation='relu')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Dropout(0.5)(x)
        x = layers.Dense(128, activation='relu')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Dropout(0.5)(x)
        outputs = layers.Dense(num_classes, activation='softmax', dtype='float32')(x)
        model = models.Model(inputs, outputs)
        optimizer = optimizers.SGD(learning_rate=0.001, momentum=0.9)

    elif model_type == 'mobilenet_frozen':
        base_model = MobileNetV2(
            input_shape=input_shape,
            include_top=False,
            weights='imagenet'
        )
        base_model.trainable = False
        model = models.Sequential([
            base_model,
            layers.GlobalAveragePooling2D(),
            layers.Dropout(0.6),
            layers.Dense(128, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.6),
            layers.Dense(num_classes, activation='softmax')
        ])
        optimizer = optimizers.Adam(learning_rate=0.0001)

    elif model_type == 'mobilenet_finetune':
        base_model = MobileNetV2(
            input_shape=input_shape,
            include_top=False,
            weights='imagenet'
        )
        for layer in base_model.layers[:-15]:
            layer.trainable = False
        model = models.Sequential([
            base_model,
            layers.MaxPooling2D((2, 2)),
            layers.Flatten(),
            layers.Dropout(0.25),
            layers.Dense(128, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.5),
            layers.Dense(num_classes, activation='softmax', dtype='float32')
        ])
        optimizer = optimizers.SGD(learning_rate=0.0001, momentum=0.9)

    elif model_type == 'paper6':
        model = models.Sequential([
            layers.Input(shape=input_shape),
            layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
            layers.BatchNormalization(),
            layers.MaxPooling2D((3, 3)),
            layers.Dropout(0.25),

            layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
            layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
            layers.BatchNormalization(),
            layers.MaxPooling2D((2, 2)),
            layers.Dropout(0.25),

            layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
            layers.BatchNormalization(),
            layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
            layers.BatchNormalization(),
            layers.MaxPooling2D((2, 2)),
            layers.Dropout(0.3),

            layers.Flatten(),
            layers.Dense(1024, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.5),
            layers.Dense(6, activation='softmax')
        ])
        optimizer = optimizers.Adam(learning_rate=0.0001)

    else:
        raise ValueError(f"Unknown model type: {model_type}. Choose 'custom', 'mobilenet', 'mobilenet_frozen', or 'mobilenet_finetune'")

    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    print("\nğŸ“Š MODEL ARCHITECTURE:")
    print("-" * 40)
    model.summary()
    total_params = model.count_params()
    trainable_params = sum([tf.keras.backend.count_params(w) for w in model.trainable_weights])
    non_trainable_params = total_params - trainable_params
    print("\nğŸ“Š MODEL STATISTICS:")
    print("-" * 40)
    print(f"Total parameters: {total_params:,}")
    print(f"Trainable parameters: {trainable_params:,}")
    print(f"Non-trainable parameters: {non_trainable_params:,}")
    print(f"Model size: ~{total_params * 4 / 1024 / 1024:.2f} MB (float32)")
    if gpu_available:
        batch_size = 16
        input_size = batch_size * 128 * 128 * 3 * 4
        param_size = total_params * 4
        activation_size = total_params * 4 * 0.5
        total_memory = (input_size + param_size + activation_size) / 1024 / 1024
        print(f"Estimated GPU memory usage: ~{total_memory:.0f} MB")
    print()

    return model

def create_callbacks():
    """Create training callbacks optimized for Mac M3 Pro GPU training"""

    print("âš™SETTING UP TRAINING CALLBACKS (MAC M3 PRO OPTIMIZED)...")
    print("-" * 40)

    # Reduce the learning rate with GPU-optimized settings
    reduce_lr = ReduceLROnPlateau(
        monitor='val_accuracy',
        factor=0.5,
        patience=10,
        min_lr=0.000001,
        verbose=1,
        mode='max',  # Explicitly set mode for accuracy
    )

    # Save the best model during training - Fixed version
    model_checkpoint = ModelCheckpoint(
        filepath=os.path.join(model_save_dir, 'best_model_m3_pro.keras'),
        monitor='val_accuracy',
        save_best_only=True,
        save_weights_only=False,
        mode='max',  # Explicitly set mode for accuracy
        verbose=1,
        initial_value_threshold=0
    )

    early_stopping = EarlyStopping(
        monitor='val_accuracy',
        patience=8,
        restore_best_weights=True,
        verbose=1
    )


    callbacks = [model_checkpoint, reduce_lr]

    print("Callbacks configured:")
    print("Model Checkpoint (best validation accuracy)")
    print()

    return callbacks

def train_model(model, train_generator, validation_generator, callbacks):
    """Train the model with Mac M3 Pro GPU acceleration"""

    print("STARTING MODEL TRAINING")
    print("=" * 70)

    # Calculate steps per epoch
    steps_per_epoch = train_generator.samples // train_generator.batch_size
    validation_steps = validation_generator.samples // validation_generator.batch_size

    # Optimized epochs for GPU training
    max_epochs = 100

    print(f"Training configuration:")
    print(f"Device: {'Mac M3 Pro GPU' if gpu_available else 'CPU'}")
    print(f"Steps per epoch: {steps_per_epoch}")
    print(f"Validation steps: {validation_steps}")
    print(f"     Batch size: {train_generator.batch_size}")
    print(f"     Mixed precision: {'Enabled' if tf.keras.mixed_precision.global_policy().name == 'mixed_float16' else 'Disabled'}")
    print()

    start_time = time.time()


    # Train the model
    history = model.fit(
        train_generator,
        epochs=max_epochs,
        steps_per_epoch=steps_per_epoch,
        validation_data=validation_generator,
        validation_steps=validation_steps,
        callbacks=callbacks,
        verbose=2,

    )

    # Calculate training time
    training_time = time.time() - start_time
    hours = int(training_time // 3600)
    minutes = int((training_time % 3600) // 60)
    seconds = int(training_time % 60)

    print(f"\nTRAINING COMPLETED!")
    print("=" * 70)
    print(f"â±Training time: {hours:02d}:{minutes:02d}:{seconds:02d}")
    print(f"Total epochs completed: {len(history.history['accuracy'])}")
    print()

    return history

def visualize_training_results(history):
    """Create comprehensive training visualizations with soft styling"""

    print("CREATING TRAINING VISUALIZATIONS")
    print("-" * 40)

    # Set up softer plotting style
    plt.style.use('seaborn-v0_8-darkgrid')  # Softer grid style
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.patch.set_facecolor('white')

    epochs_range = range(1, len(history.history['accuracy']) + 1)

    # Soft color palette
    train_color = '#2E86AB'    # Soft blue
    val_color = '#A23B72'      # Soft red/pink
    lr_color = '#F18F01'       # Soft orange

    # Training and Validation Accuracy (0 to 1 range)
    axes[0, 0].plot(epochs_range, history.history['accuracy'],
                    color=train_color, linewidth=2.5, alpha=0.8,
                    label='Training Accuracy', marker='o', markersize=3,
                    markerfacecolor=train_color, markeredgecolor='white', markeredgewidth=0.5)
    axes[0, 0].plot(epochs_range, history.history['val_accuracy'],
                    color=val_color, linewidth=2.5, alpha=0.8,
                    label='Validation Accuracy', marker='s', markersize=3,
                    markerfacecolor=val_color, markeredgecolor='white', markeredgewidth=0.5)

    axes[0, 0].set_title('Model Accuracy', fontsize=15, fontweight='bold', pad=15)
    axes[0, 0].set_xlabel('Epochs', fontsize=12)
    axes[0, 0].set_ylabel('Accuracy', fontsize=12)
    axes[0, 0].set_ylim(0, 1)  # Fixed range from 0 to 1
    axes[0, 0].legend(frameon=True, fancybox=True, shadow=True, loc='lower right')
    axes[0, 0].grid(True, alpha=0.3, linestyle='--')
    axes[0, 0].set_facecolor('#fafafa')

    # Training and Validation Loss
    axes[0, 1].plot(epochs_range, history.history['loss'],
                    color=train_color, linewidth=2.5, alpha=0.8,
                    label='Training Loss', marker='o', markersize=3,
                    markerfacecolor=train_color, markeredgecolor='white', markeredgewidth=0.5)
    axes[0, 1].plot(epochs_range, history.history['val_loss'],
                    color=val_color, linewidth=2.5, alpha=0.8,
                    label='Validation Loss', marker='s', markersize=3,
                    markerfacecolor=val_color, markeredgecolor='white', markeredgewidth=0.5)

    axes[0, 1].set_title('Model Loss', fontsize=15, fontweight='bold', pad=15)
    axes[0, 1].set_xlabel('Epochs', fontsize=12)
    axes[0, 1].set_ylabel('Loss', fontsize=12)
    axes[0, 1].legend(frameon=True, fancybox=True, shadow=True, loc='upper right')
    axes[0, 1].grid(True, alpha=0.3, linestyle='--')
    axes[0, 1].set_facecolor('#fafafa')

    # Training Summary Statistics with improved styling
    axes[1, 1].axis('off')

    # Calculate final metrics
    final_train_acc = history.history['accuracy'][-1]
    final_val_acc = history.history['val_accuracy'][-1]
    final_train_loss = history.history['loss'][-1]
    final_val_loss = history.history['val_loss'][-1]
    best_val_acc = max(history.history['val_accuracy'])

    # Status with color coding
    gap = abs(final_train_acc - final_val_acc)
    if gap < 0.05:
        status = "Excellent"
        status_color = "#2E8B57"  # Sea green
    elif gap < 0.1:
        status = "Good"
        status_color = "#FF8C00"  # Dark orange
    else:
        status = "Overfitting"
        status_color = "#DC143C"  # Crimson

    summary_text = f"""TRAINING SUMMARY
    â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

    >> Final Training Accuracy:     {final_train_acc:.4f} ({final_train_acc*100:.2f}%)
    >> Final Validation Accuracy:   {final_val_acc:.4f} ({final_val_acc*100:.2f}%)
    ** Best Validation Accuracy:    {best_val_acc:.4f} ({best_val_acc*100:.2f}%)

    >> Final Training Loss:         {final_train_loss:.4f}
    >> Final Validation Loss:       {final_val_loss:.4f}

    >> Total Epochs:                {len(epochs_range)}

    >> Overfitting Analysis:
    â”œâ”€â”€ Accuracy Gap:               {gap:.4f} ({gap*100:.2f}%)
    â””â”€â”€ Status:                     {status}
    """

    axes[1, 1].text(0.05, 0.95, summary_text, transform=axes[1, 1].transAxes,
                    fontsize=11, verticalalignment='top', fontfamily='sans-serif',
                    bbox=dict(boxstyle='round,pad=0.8', facecolor='#f8f9fa',
                              edgecolor=status_color, linewidth=2, alpha=0.9))

    # Add a subtle background gradient effect
    for ax in axes.flat:
        if ax != axes[1, 1]:  # Don't apply to summary plot
            ax.set_axisbelow(True)

    plt.tight_layout(pad=3.0)
    plt.suptitle('Training Results Overview', fontsize=18, fontweight='bold',
                 y=0.98, color='#2c3e50')

    # Save the plot with higher quality
    plot_path = os.path.join(model_save_dir, 'training_history.png')
    plt.savefig(plot_path, dpi=300, bbox_inches='tight', facecolor='white',
                edgecolor='none', format='png')
    print(f"Training plots saved to: {plot_path}")
    plt.show()

def evaluate_model(model, test_generator):
    """Comprehensive model evaluation on a test set"""

    print("EVALUATING MODEL ON TEST SET...")
    print("=" * 70)

    # Evaluate model
    test_loss, test_accuracy = model.evaluate(test_generator, verbose=1)

    print(f"\nTEST RESULTS:")
    print(f"     Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
    print(f"     Test Loss: {test_loss:.4f}")

    # Generate predictions for detailed analysis
    print("\nGENERATING DETAILED PREDICTIONS...")

    # Get predictions
    test_generator.reset()
    predictions = model.predict(test_generator, verbose=1)
    predicted_classes = np.argmax(predictions, axis=1)

    # Get true labels
    true_classes = test_generator.classes
    class_labels = list(test_generator.class_indices.keys())

    # Classification report
    print("\nCLASSIFICATION REPORT:")
    print("=" * 70)
    report = classification_report(true_classes, predicted_classes,
                                   target_names=class_labels, digits=4)
    print(report)

    # Save a classification report
    report_path = os.path.join(model_save_dir, 'classification_report.txt')
    with open(report_path, 'w') as f:
        f.write("SKIN LESION CLASSIFICATION REPORT\n")
        f.write("=" * 50 + "\n\n")
        f.write(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)\n")
        f.write(f"Test Loss: {test_loss:.4f}\n\n")
        f.write("DETAILED CLASSIFICATION REPORT:\n")
        f.write(report)

    print(f"Classification report saved to: {report_path}")

    # Confusion Matrix
    create_confusion_matrix(true_classes, predicted_classes, class_labels)

    return test_accuracy, test_loss

def create_confusion_matrix(true_classes, predicted_classes, class_labels):
    """Create and display confusion matrix"""

    print("\nCREATING CONFUSION MATRIX...")

    # Calculate confusion matrix
    cm = confusion_matrix(true_classes, predicted_classes)

    # Create a figure
    plt.figure(figsize=(10, 8))

    # Create heatmap
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_labels, yticklabels=class_labels,
                cbar_kws={'label': 'Number of Samples'})

    plt.title('Confusion Matrix - Skin Lesion Classification',
              fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Predicted Labels', fontsize=12)
    plt.ylabel('True Labels', fontsize=12)

    # Add accuracy per class
    class_accuracies = cm.diagonal() / cm.sum(axis=1)
    for i, acc in enumerate(class_accuracies):
        plt.text(len(class_labels) + 0.5, i + 0.5, f'{acc:.3f}',
                 ha='center', va='center', fontweight='bold', color='red')

    plt.tight_layout()

    # Save confusion matrix
    cm_path = os.path.join(model_save_dir, 'confusion_matrix.png')
    plt.savefig(cm_path, dpi=300, bbox_inches='tight')
    print(f"Confusion matrix saved to: {cm_path}")
    plt.show()

    # Print per-class accuracy
    print("\nğŸ“Š PER-CLASS ACCURACY:")
    print("-" * 40)
    for i, (label, acc) in enumerate(zip(class_labels, class_accuracies)):
        print(f"     {label}: {acc:.4f} ({acc*100:.2f}%)")

def save_model_info(model, history, test_accuracy):
    """Save comprehensive model information"""

    print("\nSAVING MODEL INFORMATION...")

    # Save model architecture
    model_json = model.to_json()
    json_path = os.path.join(model_save_dir, 'model_architecture.json')
    with open(json_path, 'w') as json_file:
        json_file.write(model_json)

    # Save model weights
    weights_path = os.path.join(model_save_dir, 'model.weights.h5')
    model.save_weights(weights_path)

    # Save a complete model
    full_model_path = os.path.join(model_save_dir, 'complete_model.keras')
    model.save(full_model_path)

    # Save training history
    history_df = pd.DataFrame(history.history)
    history_path = os.path.join(model_save_dir, 'training_history.csv')
    history_df.to_csv(history_path, index=False)

    # Create a model summary file
    summary_path = os.path.join(model_save_dir, 'model_summary.txt')
    with open(summary_path, 'w') as f:
        f.write("SKIN LESION CLASSIFICATION MODEL SUMMARY\n")
        f.write("=" * 50 + "\n\n")
        f.write(f"Model Type: Convolutional Neural Network (CNN)\n")
        f.write(f"Input Shape: (150, 150, 3)\n")
        f.write(f"Number of Classes: 6\n")
        f.write(f"Total Parameters: {model.count_params():,}\n")
        f.write(f"Training Epochs: {len(history.history['accuracy'])}\n")
        f.write(f"Best Validation Accuracy: {max(history.history['val_accuracy']):.4f}\n")
        f.write(f"Final Test Accuracy: {test_accuracy:.4f}\n\n")
        f.write("Classes:\n")
        f.write("- ACK: Actinic Keratosis\n")
        f.write("- BCC: Basal Cell Carcinoma\n")
        f.write("- MEL: Melanoma\n")
        f.write("- NEV: Nevus\n")
        f.write("- SEK: Seborrheic Keratosis\n")
        f.write("- SCC: Squamous Cell Carcinoma\n\n")
        f.write("Files Saved:\n")
        f.write(f"- Complete Model: {full_model_path}\n")
        f.write(f"- Model Weights: {weights_path}\n")
        f.write(f"- Model Architecture: {json_path}\n")
        f.write(f"- Training History: {history_path}\n")

    print("Model information saved:")
    print(f"Model summary: {summary_path}")
    print(f"Complete model: {full_model_path}")
    print(f"Model weights: {weights_path}")
    print(f"Model architecture: {json_path}")
    print(f"Training history: {history_path}")

def main():
    """Main execution function"""

    # Verify dataset exists
    if not verify_dataset():
        print("TRAINING ABORTED: Dataset verification failed")
        return

    # Create data generators
    train_gen, val_gen, test_gen = create_data_generators()

    # Create model
    model = create_model(model_type='mobilenet_finetune', num_classes=6, input_shape=(96, 96, 3))
    # Setup callbacks
    callbacks = create_callbacks()

    # Train model
    history = train_model(model, train_gen, val_gen, callbacks)

    # Load the best model for evaluation
    best_model_path = os.path.join(model_save_dir, 'best_model_m3_pro.keras')
    if os.path.exists(best_model_path):
        print("Loading best model for evaluation...")
        model = tf.keras.models.load_model(best_model_path)


    # Visualize training results
    visualize_training_results(history)

    # Evaluate model
    test_accuracy, test_loss = evaluate_model(model, test_gen)

    # Save model and information
    save_model_info(model, history, test_accuracy)

    # Final summary
    print("\nMODEL TRAINING COMPLETED SUCCESSFULLY")
    print("=" * 70)

    print(f"ğŸ¯ Final Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
    print(f"ğŸ“ All results saved in: {model_save_dir}")


if __name__ == "__main__":
    # Set random seeds for reproducibility
    np.random.seed(42)
    tf.random.set_seed(42)

    # Run the main training pipeline
    main()

```
